# Exploratory Data Analysis (EDA)

This notebook explores the structure and basic characteristics of customer review data
used for aspect-based sentiment analysis.


```python
import pandas as pd

data = {
    "review_id": [1, 2, 3],
    "review_text": [
        "The battery life is excellent but the screen quality is poor.",
        "Customer service was very helpful and friendly.",
        "Delivery was slow, but the product quality is good."
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["review_length"] = df["review_text"].str.split().str.len()
df["review_length"].describe()
```

### Initial Observations
- Reviews frequently contain multiple aspects
- This supports the need for aspect-based sentiment analysis
