# Modeling Overview: Aspect-Based Sentiment Analysis

This notebook provides a high-level overview of the modeling approach used in the project,
focusing on aspect extraction and aspect-level sentiment classification.


## Modeling Approach

The project follows a two-stage Aspect-Based Sentiment Analysis (ABSA) pipeline:

1. **Aspect Extraction**
   - Fine-tuned BERT model
   - Identifies product aspects such as features and services mentioned in reviews

2. **Aspect-Level Sentiment Classification**
   - Fine-tuned RoBERTa model
   - Classifies sentiment (positive, negative, neutral) associated with each extracted aspect


```python
# Example structure of aspect-sentiment predictions

example_predictions = [
    {"aspect": "battery life", "sentiment": "positive"},
    {"aspect": "screen quality", "sentiment": "negative"},
    {"aspect": "customer service", "sentiment": "positive"}
]

example_predictions
```

- **RoBERTa (Sentiment Classification)**
